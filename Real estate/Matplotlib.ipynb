```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[2,4,5])
plt.title("This is data")
plt.ylabel("this is y")
plt.xlabel("this is x")
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x=[1,2,3,4]
y=[3,4,5,6]
x1=[4,5,6,7]
y1=[6,7,8,9]
x2=[1,2,3,4]
y2=[9,8,7,6]
x3=[4,6,7]
y3=[6,4,3]

plt.plot(x,y , label="first", linewidth=4)
plt.plot(x1,y1 , label="second")
plt.plot(x2,y2 , label="third")
plt.plot(x3,y3 , label="fourth")
plt.legend()
plt.show()
```
